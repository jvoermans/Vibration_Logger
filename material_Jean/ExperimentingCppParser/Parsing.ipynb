```c++
#include <fstream>
#include <iterator>
#include <vector>
#include <iostream>
#include <cstdlib>
```

```c++
auto path_to_file = "/home/jrlab/Desktop/Git/Vibration_Logger/material_Jean/BinarySdDataParser/example_data/F00000000.bin";
std::ifstream input{path_to_file, std::ios::binary};
```

```c++
std::vector<unsigned char> buffer{std::istreambuf_iterator<char>(input), {}};
```

```c++
char cout_buff [3];

for (auto const & elem : buffer){
    sprintf(cout_buff, "%02X", elem);
    std::cout << cout_buff << " ";
}

std::cout << std::endl;
```
